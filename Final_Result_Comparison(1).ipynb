```python
# Import Library
import numpy as np 
import matplotlib.pyplot as plt 

# Define Data
data3 = ['LIAR Dataset','ISOT Dataset','Twitter Dataset','Kaggle Dataset']
xvals = [58.09,98.73,93.3,94.87]
yvals = [61.48,99.57,87.41,89.36]
zvals = [54.22,99.21,93.75,90.87]
avals = [62.83,94.65,90.51,84.66]
bvals = [58.25,99.25,94.03,96.58]

x_axis = np.arange(len(data3))

# Multi bar Chart
plt.bar(x_axis +0.1, xvals, width=0.1, label = 'Logistic Regression')
plt.bar(x_axis +0.1*2, yvals, width=0.1, label = 'Decision tree')
plt.bar(x_axis +0.1*3, zvals, width=0.1, label = 'Random Forest')
plt.bar(x_axis +0.1*4, avals, width=0.1, label = 'Naive Bayes')
plt.bar(x_axis +0.1*5, bvals, width=0.1, label = 'SVM')

# Xticks
plt.xticks(x_axis,data3)
plt.ylabel("Accuracy (%)")
plt.xlabel("Experiments on Different datasets")
plt.title("Comparitive analysis of different datasets using ML Algorithms")
plt.legend()
plt.show()
```
